```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Define fuzzy variables
# Input variables: Temperature (°C) and Humidity (%)
temperature = ctrl.Antecedent(np.arange(16, 36, 1), 'temperature')
humidity = ctrl.Antecedent(np.arange(20, 81, 1), 'humidity')

# Output variable: Fan Speed (0-100%)
fan_speed = ctrl.Consequent(np.arange(0, 101, 1), 'fan_speed')

# Membership functions for Temperature
temperature['cold'] = fuzz.trapmf(temperature.universe, [16, 16, 20, 24])
temperature['cool'] = fuzz.trimf(temperature.universe, [20, 24, 28])
temperature['warm'] = fuzz.trimf(temperature.universe, [24, 28, 32])
temperature['hot'] = fuzz.trapmf(temperature.universe, [28, 32, 36, 36])

# Membership functions for Humidity
humidity['low'] = fuzz.trapmf(humidity.universe, [20, 20, 30, 40])
humidity['medium'] = fuzz.trimf(humidity.universe, [30, 50, 70])
humidity['high'] = fuzz.trapmf(humidity.universe, [60, 70, 80, 80])

# Membership functions for Fan Speed
fan_speed['low'] = fuzz.trapmf(fan_speed.universe, [0, 0, 20, 40])
fan_speed['medium'] = fuzz.trimf(fan_speed.universe, [20, 50, 80])
fan_speed['high'] = fuzz.trapmf(fan_speed.universe, [60, 80, 100, 100])

# Define fuzzy rules
rule1 = ctrl.Rule(temperature['cold'] & humidity['low'], fan_speed['low'])
rule2 = ctrl.Rule(temperature['cold'] & humidity['medium'], fan_speed['low'])
rule3 = ctrl.Rule(temperature['cold'] & humidity['high'], fan_speed['medium'])

rule4 = ctrl.Rule(temperature['cool'] & humidity['low'], fan_speed['low'])
rule5 = ctrl.Rule(temperature['cool'] & humidity['medium'], fan_speed['medium'])
rule6 = ctrl.Rule(temperature['cool'] & humidity['high'], fan_speed['medium'])

rule7 = ctrl.Rule(temperature['warm'] & humidity['low'], fan_speed['medium'])
rule8 = ctrl.Rule(temperature['warm'] & humidity['medium'], fan_speed['medium'])
rule9 = ctrl.Rule(temperature['warm'] & humidity['high'], fan_speed['high'])

rule10 = ctrl.Rule(temperature['hot'] & humidity['low'], fan_speed['medium'])
rule11 = ctrl.Rule(temperature['hot'] & humidity['medium'], fan_speed['high'])
rule12 = ctrl.Rule(temperature['hot'] & humidity['high'], fan_speed['high'])

# Create control system
fan_speed_control = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12])
fan_speed_simulation = ctrl.ControlSystemSimulation(fan_speed_control)

# Example input
input_temperature = 30  # °C
input_humidity = 65  # %

# Set inputs
fan_speed_simulation.input['temperature'] = input_temperature
fan_speed_simulation.input['humidity'] = input_humidity

# Compute the output
fan_speed_simulation.compute()

# Output the result
print(f"Input Temperature: {input_temperature}°C")
print(f"Input Humidity: {input_humidity}%")
print(f"Fan Speed: {fan_speed_simulation.output['fan_speed']:.2f}%")
```
